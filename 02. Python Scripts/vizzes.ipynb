# 01. Importing Libraries

```python
import os
import math
import random
import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns
import scipy
import numpy as np
import pandas as pd
```

# 02. Importing data

```python
avg_age_adopted = pd.read_csv('../01. Data/03. Tables for visualization/adoption_avg_age.csv')
```

```python
avg_age_all_animals = pd.read_csv('../01. Data/03. Tables for visualization/avg_age_all_animals.csv')
```

```python

avg_age_shelter = pd.read_csv('../01. Data/03. Tables for visualization/avg_age_shelter_animals.csv')
```

# 03. Visualizations

```python
# defining color dictionary for consistent use across all visualizations
colors = {
    "primary":    "#f55e61",  # coral red — main colour
    "secondary":  "#f9a8a9",  # light pink — secondary bars/elements
    "dark":       "#2a0a0b",  # deep dark red — backgrounds/text
    "cream":      "#FFF5F5",  # warm white — plot background
    "muted":      "#9a6060",  # muted rose — gridlines/labels
    "accent":     "#d94a4d",  # darker red — highlights
}
```

#### Dataframe sorted creation

```python
avg_age_adopted_sorted = avg_age_adopted.sort_values('avg_age_yo', ascending=False)
```

```python
avg_age_all_animals_sorted = avg_age_all_animals.sort_values('avg_age_yo', ascending=False)
```

```python
avg_age_shelter_sorted = avg_age_shelter.sort_values('avg_age_yo', ascending=False)
```

#### Creation of visualizations

```python
# barplot on average age of adopted animals

adopted = sns.barplot(data=avg_age_adopted, x='pet_type', y='avg_age_yo', palette=[colors["muted"]])


adopted.set_title('Average Age of Adopted Animals')
adopted.set_xlabel('Pet Type')
adopted.set_ylabel('Average Age (years)')
adopted.tick_params(axis='x', rotation=45)

plt.tight_layout() 
plt.show()
```

```python
# exporting the figure
adopted.get_figure().savefig('../04. Visualizations/avg_age_adopted.png', dpi=300, bbox_inches='tight')
```

```python
# barplot on average age of adopted animals, sorted by avg age

adopted_sorted = sns.barplot(data=avg_age_adopted_sorted, x='pet_type', y='avg_age_yo', order=avg_age_adopted_sorted['pet_type'], palette=[colors["muted"]])

adopted_sorted.set_title('Average Age of Adopted Animals (Sorted by Age)')
adopted_sorted.set_xlabel('Pet Type')
adopted_sorted.set_ylabel('Average Age (years)')
adopted_sorted.tick_params(axis='x', rotation=45)

plt.tight_layout() 
plt.show()
```

```python
# barplot on average age all animals

all_animals = sns.barplot(data=avg_age_all_animals, x='pet_type', y='avg_age_yo', palette=[colors["muted"]])


all_animals.set_title('Average Age of All Animals')
all_animals.set_xlabel('Pet Type')
all_animals.set_ylabel('Average Age (years)')
all_animals.tick_params(axis='x', rotation=45)

plt.tight_layout() 
plt.show()
```

```python
# barplot on average age of all animals, sorted by avg age

all_animals_sorted = sns.barplot(data=avg_age_all_animals_sorted, x='pet_type', y='avg_age_yo', order=avg_age_all_animals_sorted['pet_type'], palette=[colors["muted"]])

all_animals_sorted.set_title('Average Age of All Animals (Sorted by Age)')
all_animals_sorted.set_xlabel('Pet Type')
all_animals_sorted.set_ylabel('Average Age (years)')
all_animals_sorted.tick_params(axis='x', rotation=45)

plt.tight_layout() 
plt.show()
```

```python
# barplot on average age non adopted animals

shelter = sns.barplot(data=avg_age_shelter, x='pet_type', y='avg_age_yo', palette=[colors["muted"]])


shelter.set_title('Average Age of Shelter Animals')
shelter.set_xlabel('Pet Type')
shelter.set_ylabel('Average Age (years)')
shelter.tick_params(axis='x', rotation=45)

plt.tight_layout() 
plt.show()
```

```python
# barplot on average age non adopted animals sorted by age
shelter_sorted = sns.barplot(data=avg_age_shelter_sorted, x='pet_type', y='avg_age_yo', palette=[colors["muted"]])

shelter_sorted.set_title('Average Age of Shelter Animals (Sorted by Age)')
shelter_sorted.set_xlabel('Pet Type')
shelter_sorted.set_ylabel('Average Age (years)')
shelter_sorted.tick_params(axis='x', rotation=45)

plt.tight_layout() 
plt.show()
```
