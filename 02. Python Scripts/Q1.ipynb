```python
import pandas as pd
import matplotlib.pyplot as plt

black_cats_other_cats = pd.read_csv("black_cats_other_cats.csv")
count_cats_by_color = pd.read_csv("Count_cats_by_color.csv")

# Chart 1
top_counts = count_cats_by_color.sort_values("total_cats", ascending=False).head(15)

fig1, ax1 = plt.subplots(figsize=(12, 6))
ax1.bar(top_counts["color"], top_counts["total_cats"], color="#D94A4D")
ax1.set_xlabel("Color")
ax1.set_ylabel("Number of Cats")
ax1.set_title("How Many Cats of Each Color the Shelter Has")
ax1.tick_params(axis="x", rotation=45)
ax1.spines["top"].set_visible(False)
ax1.spines["right"].set_visible(False)
ax1.spines["left"].set_visible(False)
ax1.spines["bottom"].set_visible(False)
ax1.set_frame_on(False)
fig1.tight_layout()
fig1.savefig("cats_by_color.pdf", format="pdf", bbox_inches="tight")
plt.show()

# Chart 2
fig2, ax2 = plt.subplots(figsize=(8, 5))
ax2.bar(black_cats_other_cats["color_group"], black_cats_other_cats["adoption_rate_pct"], color=["#D94A4D", "#2E2E2E"])
ax2.set_xlabel("Color Group")
ax2.set_ylabel("Adoption Rate (%)")
ax2.set_title("Black Cats vs Other Cats Adoption Rate")
ax2.spines["top"].set_visible(False)
ax2.spines["right"].set_visible(False)
ax2.spines["left"].set_visible(False)
ax2.spines["bottom"].set_visible(False)
ax2.set_frame_on(False)
fig2.tight_layout()
fig2.savefig("black_vs_other.pdf", format="pdf", bbox_inches="tight")
plt.show()
```
